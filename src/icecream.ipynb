```python
import pandas as pd
import matplotlib.pyplot as plt

#Criação da Variável correspondente ao dataset original#
dado = pd.read_csv('IceCreamData.csv')
#Organização do resultado em ordem crescente de temperatura#
dado = dado.sort_values(['Temperature'])
#Criação do gráfico já com a definição de cor, marcador e delimitador bem como com a legenda#
plt.plot(dado['Temperature'], dado['Revenue'], 'r.--', label = 'US$')
#Definição do título do gráfico#
plt.title('Caminhão de Sorvetes')
#Definição dos nomes dos eixos#
plt.xlabel('Temperatura')
plt.ylabel('Vendas em US$')
#Aplicação da legenda
plt.legend()
#Criação do arquivo .pdf contendo o gráfico referente ao 
plt.savefig('resultado.pdf')
```
